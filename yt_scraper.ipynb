```python
import time

user_data = []
links = []
count =0

#set up the driver to fetch the content of url from YouTube
driver = webdriver.Chrome('C:/Users/nitin/OneDrive/Desktop/Projects/FR_With_Siamese/chromedriver_win32/chromedriver')

#open the text file having the keywords and fetch the page
with open('C:/Users/nitin/OneDrive/Desktop/Projects/FR_With_Siamese/keywords.txt','r') as file:
    keywords = file.readlines()
    for keyword in keywords:
        driver.get("https://www.youtube.com/results?search_query="+keyword)
        time.sleep(5)
        
        #fetch the path to the desired objects
        user_data = driver.find_elements_by_xpath('//*[@id="video-title"]')
        
        for i in user_data:
            count+=1
            #fetch the urls out of the objects fetched
            links.append(i.get_attribute('href'))
            if count>=5:
                count = 0
                break
print(len(links))
driver.close()
```

```python
# Program To Read video 
# and Extract Frames 
import os
import cv2 
  
# Function to extract frames 
def FrameCapture(path): 
      
    # Path to video file 
    vidObj = cv2.VideoCapture(path) 
  
    # Used as counter variable 
    count = 0
  
    # checks whether frames were extracted 
    success = 1
  
    while success: 
  
        # vidObj object calls read 
        # function extract frames 
        success, image = vidObj.read()
        
        #save every 150th frame
        if count%150 == 0:
            print(count)
            print('here')
            # Saves the frames with frame-count
            for root, dirs, files in os.walk("C:/Users/nitin/OneDrive/Desktop/Projects/FR_With_Siamese/test", topdown=False):
                i = 0
                for Dir in dirs:
                    for cwd, dirs, files in os.walk(os.path.join(root,Dir), topdown=False):
                        for file in files:
                            Output = 'C:/Users/nitin/OneDrive/Desktop/Projects/FR_With_Siamese/Output'
                            print(Output)
                            cv2.imwrite(Output+os.path.sep+str(i)+os.path.sep+str(path.split('.')[0])+"_"+"%d.jpg" % count, image)
                            print(i)
                    i += 1
  
        count += 1

    
# Driver Code 
if __name__ == '__main__': 
  
    # Calling the function
    for root, dirs, files in os.walk("C:/Users/nitin/OneDrive/Desktop/Projects/FR_With_Siamese/test/", topdown=False):
        for Dir in dirs:
            for cwd, dirs, files in os.walk(os.path.join(root,Dir), topdown=False):
                for file in files:
                    cur_file = os.path.join(cwd,file)
                    FrameCapture(cur_file) 
```
